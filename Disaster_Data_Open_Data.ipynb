```python
#1971-2010 data from open data nepal
# merging two csv files
df = pd.concat(
	map(pd.read_csv, ['data/opendata_previous.csv', 'data/opendata_latest.csv']), ignore_index=True)
df
```

```python
df
```

```python
df.rename(columns={'district': 'District', 'Geographical Region': 'Geographical_Region',
                  'region': 'Region',
                  'village': 'Village', 
                   'Data Cards': 'Data_Cards',
                  'Houses Destroyed': 'Houses_Destroyed',
                   'Houses Damaged': 'Houses_Damaged',
                  'Losses USD': 'Losses_USD',
                'Education centers': 'Education_Centers','Damages in crops Ha.': 'Damages__Crops_Ha',
                   'Losses Local': 'Losses_Local'}, inplace=True)
```

```python
df['Data_Cards'] = df['Data_Cards'].fillna(df['Data_Cards'].value_counts().idxmax())
```

```python
df['Data_Cards'].isna().sum()
```

```python
df['Source'] = df['Source'].fillna(df['Source'].value_counts().idxmax())
```

```python
df['Source'].isna().sum()
```

```python
df['Village'] = df.apply(lambda row: row['District'] if pd.isnull(row['Village']) else row['Village'], axis=1)
```
